```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import iv, jv
```

```python
# Prepare x-axis values (start close to 0 and cover a wide range) (FIgure 4)
x = np.linspace(0.01, 10, 400)

# Set parameters sigma and m
sigma = -0.0001
m = 0.5

# Variable adjustment
x_sigma_sqrt = x * np.sqrt(-sigma)

# Modified first kind Bessel function I_1(x*sqrt(sigma)) / (x*sqrt(sigma))
I1_x_sigma_sqrt = 2 * iv(1, x_sigma_sqrt) / x_sigma_sqrt

# Exponential function y = e^{-x * sigma}
exp_neg_x_sigma = np.exp(-x * sigma)

# Exponential function y = e^(-x * sigma / (1-m)}
exp_neg_x_sigma_m = np.exp(-x * sigma / (1 - m))

# Plotting the graph
plt.figure(figsize=(8, 5))

plt.plot(x, I1_x_sigma_sqrt, label=r'$\frac{2 I_1(x\sqrt{\sigma})}{x\sqrt{\sigma}}$')
plt.plot(x, exp_neg_x_sigma, '--', label=r'$e^{-x \sigma}$', color='black')
plt.plot(x, exp_neg_x_sigma_m, ':', label=r'$e^{-\frac{x \sigma}{1-m}}$', color='red')

plt.xlabel('x')
plt.ylabel('Function value', fontsize=14)
plt.yticks(np.arange(1.000, 1.002, 0.001), fontsize=14)

plt.legend(fontsize=18)

plt.savefig('comparison_functions.pdf')
plt.show()
```

```python
# Prepare x-axis values (start close to 0 and cover a wide range) (Figure 4)
x = np.linspace(0.01, 50, 400)

# Set parameters sigma and m
sigma = 0.04
m = 0.5

# Variable adjustment
x_sigma_sqrt = x * np.sqrt(sigma)

# Multiply the ordinary first kind Bessel function J_1(x*sqrt(sigma)) / (x*sqrt(sigma)) by 2
J1_x_sigma_sqrt = 2 * jv(1, x_sigma_sqrt) / x_sigma_sqrt

# Exponential function y = e^{-x * sigma}
exp_neg_x_sigma = np.exp(-x * sigma)

# Exponential function y = e^{-x * sigma / (1-m)}
exp_neg_x_sigma_m = np.exp(-x * sigma / (1 - m))

# Plotting the graph
plt.figure(figsize=(8, 5))

plt.plot(x, J1_x_sigma_sqrt, label=r'$2 \cdot \frac{J_1(x\sqrt{\sigma})}{x\sqrt{\sigma}}$')
plt.plot(x, exp_neg_x_sigma, '--', label=r'$e^{-x \sigma}$', color='black')
plt.plot(x, exp_neg_x_sigma_m, ':', label=r'$e^{-\frac{x \sigma}{1-m}}$', color='red')

plt.xlabel('x')
plt.ylabel('Function value', fontsize=14)
plt.yticks(np.arange(0, 1.002, 0.2), fontsize=14)
plt.legend(fontsize=18)

plt.savefig('comparison_functions_positive_eigenvalue.pdf')
plt.show()
```
